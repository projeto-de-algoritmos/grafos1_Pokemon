```python
import networkx as nx
G = nx.Graph()

# Adicionando o nó
G.add_node(1)


# Visualizando os nós
print(G.nodes())


# Adicionando arestas
G.add_edge('arthur', 'mayra')
G.add_edge('arthur', 'lua')
G.add_edge('arthur', 'pérola')
G.add_edge('arthur', 'alice')
G.add_edge('arthur', 'giovanna')
G.add_edge('arthur', 'bruna')
G.add_edge('arthur', 'camila')
G.add_edge('arthur', 'may')
G.add_edge('arthur', 'aricia')
G.add_edge('arthur', 'carboidrato')


# Adicioanndo relacionamento entre as arestas
G.add_edge('perola', 'lua')
G.add_edge('lua', 'alice')
G.add_edge('alice', 'giovanna')
G.add_edge('giovanna', 'bruna')
G.add_edge('bruna', 'camila')
G.add_edge('camila', 'may')
G.add_edge('may', 'mayra')
G.add_edge('mayra', 'aricia')


# Total de vertices e arestas
n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()

# Print com o número de vértices e arestas 
print('vertices: ', n_vertices, '\narestas: ', n_arestas)
```

Removendo um nó que não está sendo utilizado (por algum motivo fica aparecendo esse nó aleatório quando imprime o grafo)

```python
G.remove_node(1)
print(G.nodes())
```

Impressão do grafo

```python
nx.draw(G)
```

Impressão do grafo com os nomes dos envolvidos

```python
nx.draw(G, with_labels=True)
```

Impressão dos nós que não foram visitados, e por esse motivo eles ficam em vermelho

```python
nx.draw(G, with_labels=True, node_size=1200, node_color='red')
```

Construção de um subgrafo 

```python
subgrafos = [G.subgraph(s) for s in nx.connected_components(G)]

print(len(subgrafos)) 
```

O grafo apresentado em vermelho é o 'grafo principal', que foi apresentado acima e que apresenta os envolvidos.

O grafo em vermelho é o subgrafo que irá apresenta as conexoes entre todos os envolvidos.


```python
# grafo vermelho - circular
nx.draw_circular(
    subgrafos[0], 
    with_labels=True, 
    node_size=800, 
    node_color='red'
)

# grafo verde - posições aleatórias
nx.draw_random(
    subgrafos[0], 
    with_labels=True, 
    node_size=800, 
    node_color='green'
)
```
